```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd

def projectileMotion(h,u,theta,t):
    g = 10
    ux = u * math.cos(theta / 180 * math.pi)
    uy = u * math.sin(theta / 180 * math.pi)
    tmax = (uy + math.sqrt(uy**2 + 2*g*h)) / g
    if t >= tmax:
        t = tmax
    sx = ux * t
    sy = h + uy * t - 0.5 * g * t ** 2
    return sx, sy
```

```python
times = [i * 0.1 for i in range(0, 100)]
h = 50
u = 30
theta = 45

sx_all = []
sy_all = []
```

```python

for t in times:
    plt.clf()
    plt.axis([-10, 250, -10, 150])
    plt.grid()
    plt.plot([-10,0,0,250],[h,h,0,0],'b-')    
    sx,sy = projectileMotion(h,u,theta,t)
    sx_all.append(sx)
    sy_all.append(sy)
    plt.plot(sx_all,sy_all,'k.',lw=0)
    plt.plot(sx,sy,'ro')
    plt.pause(0.1)
    
plt.show()
```

```python
spending_record = [[100,150,80,200,70,390,50,50,180,70,130,50,40,65,40],
                   [100,120,60,0,30,150,200,120,100,60,45,30,0,0,20],
                   [430,0,0,100,0,0,200,120,350,0,0,0,100,0,35],
                   [0,25,40,100,120,100,80,0,20,10,200,10,40,250,50]]

types = ['food','transport','shopping','others']
```

```python
# heatmap
fig, ax = plt.subplots()
im = ax.imshow(spending_record)

ax.set_xticks(np.arange(len(spending_record[0])))
ax.set_yticks(np.arange(len(types)))
ax.set_xticklabels(np.arange(1,len(spending_record[0])+1))
ax.set_yticklabels(types)
ax.set_xlabel('Day')
```

```python
# horizontal bar
a = np.array(spending_record)
fig, ax = plt.subplots()
ax.barh(types,a.sum(axis=1))
ax.invert_yaxis()
ax.set_xlabel('Spending (baht)')

# palette = plt.get_cmap('Set1')
# color=palette(num)
```

```python
# line graph
x = np.tile(np.arange(len(spending_record[0])),(len(types)+1,1))
labels = types[::]
labels.append('total')
df = pd.DataFrame(np.append(a,[a.sum(axis=0)],axis=0).T,columns=labels)
df.index += 1
ax = df.plot()
ax.set_xticks(np.arange(1,len(spending_record[0])+1))
ax.set_ylabel('Spending (baht)')
ax.set_xlabel('Day')
ax.legend()
```

```python
# pie 
sizes = a.sum(axis=1)
sizes = sizes / sizes.sum() * 100
fig, ax = plt.subplots()
ax.pie(sizes, labels=types, autopct='%1.2f%%')
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


```
